```python
%matplotlib inline
from matplotlib.pyplot import imshow
from matplotlib.pyplot import cm
import custom
import numpy as np
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier,RandomTreesEmbedding
from sklearn.datasets import load_iris,load_digits
from ggplot import *
import pandas as pd
from sklearn.metrics import euclidean_distances
from joblib import Parallel,delayed
```

```python
def dist(x):
    new = custom.reconstruct(re,re.apply(x)[0],64,0,16)
    return euclidean_distances(new,x)[0][0]
```

```python
digit = load_digits()
X = digit.data
Y = digit.target
test = digit.data[0]
```

```python
df = pd.DataFrame()
```

```python
for n in range(10,110,20):
    cls = ExtraTreesClassifier
    max_depth = None

    re = cls(max_depth=max_depth,n_estimators=n)
    re.fit(X,Y)
    l = Parallel(n_jobs=3)(delayed(dist)(x) for x in X)
    df = pd.concat([pd.DataFrame({"error":l,"n":n}),df])   
```

```python
ggplot(aes(x="error",color="n"),data=df)+geom_density()+scale_color_gradient(low='#05D9F6', high='#5011D1')
```

```python
df = pd.DataFrame()
for n in range(10,110,20):
    cls = RandomForestClassifier
    max_depth = None

    re = cls(max_depth=max_depth,n_estimators=n,bootstrap=False,max_features=1,n_jobs=-1)
    re.fit(X,Y)
    l = Parallel(n_jobs=3)(delayed(dist)(x) for x in X)
    df = pd.concat([pd.DataFrame({"error":l,"n":n}),df])   
```

```python
ggplot(aes(x="error",color="n"),data=df)+geom_density()+scale_color_gradient(low='#05D9F6', high='#5011D1')
```
