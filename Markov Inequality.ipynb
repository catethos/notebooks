**Markov Inequality:**

Let $X$ be a non-negative random variable and $a$ a non-negative real number, then the following inequality hold.

$$
P( X \geq a ) \leq \frac{\mathbb{E}( X )}{a}
$$

**Proof:**

\begin{align*}
\mathbb{E}(X) & = \sum_x xP(X=x)\\
              & = \sum_{x<a} xP(X=x) + \sum_{x \geq a} xP(X=x)\\
              & \geq \sum_{x \geq a} xP(X=x)\\
              & \geq a \sum_{x \geq a} P(X=x)\\
              & = aP(X \geq a)
\end{align*}
