```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.ensemble import RandomForestClassifier
%matplotlib inline
import matplotlib.pyplot as plt
import pickle

def contain_in(x,y):
    for u,v in zip(x,y):
        if not (u in v):
            return False
    return True

def error(X,cls):
    np.random.shuffle(X)
    #X_500 = (set(x) for x in cls.apply(X[:500]).T)
    X_500 = cls.apply(X[:500])
    error = 0
    for t in cls.apply(X[500:]):
        if contain_in(t,X_500.T):
            error+=1
    return error
```

```python
with open("error.pkl","rb") as f:
    error1000 = pickle.load(f)
with open("error2000.pkl","rb") as f:
    error2000 = pickle.load(f)
with open("error3000.pkl","rb") as f:
    error3000 = pickle.load(f)
with open("error4000.pkl","rb") as f:
    error4000 = pickle.load(f)
```

```python
n_trees = ["200","300","400","500","600"]
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['right'].set_color('none')
ax.tick_params(labelcolor='w', top='off', bottom='off', left='off', right='off')
ax.set_ylabel("Normalized Frequency",fontsize=15)
ax.set_xlabel("Error Rate(%)",fontsize=15)

ax1 = fig.add_subplot(411)
for k in n_trees:
    ax1.hist(error1000[k]/12.97,alpha=0.3,bins=10,normed=True,label=k)
ax1.legend()  
ax1.set_title("number of training data: 1000")
ax1.xaxis.set_visible(False)

ax2 = fig.add_subplot(412,sharex=ax1,sharey=ax1)
for k in n_trees:
    ax2.hist(error2000[k]/12.97,alpha=0.3,bins=10,normed=True,label=k)
ax2.set_title("number of training data: 2000")
ax2.xaxis.set_visible(False)

ax3 = fig.add_subplot(413,sharex=ax1,sharey=ax1)
for k in n_trees:
    ax3.hist(error3000[k]/12.97,alpha=0.3,bins=10,normed=True,label=k)
ax3.xaxis.set_visible(False)
ax3.set_title("number of training data: 3000")

ax4 = fig.add_subplot(414,sharex=ax1,sharey=ax1)
for k in n_trees:
    ax4.hist(error4000[k]/12.97,alpha=0.3,bins=10,normed=True,label=k)
ax4.set_title("number of training data: 4000")    
fig.savefig("histogram.png")
plt.show()
```
